```python
from state import State

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
app_state = State()

w_ball = []
w_flower = []
w_baby = []
for i in range(200):
    context = app_state.context
    toys = context.toys
    ball = next((x for x in toys if x.name == '공'), None)
    w_ball.append(ball.weight)

    flower = next((x for x in toys if x.name == '꽃'), None)
    w_flower.append(flower.weight)

    baby = next((x for x in toys if x.name == '아기'), None)
    w_baby.append(baby.weight)

    if i < 100:
        context.on_visual_attention('공')
    else:
        context.on_visual_attention('꽃')
```

```python
plt.figure(figsize=(5,3), dpi=150)
plt.title('Current with $\\alpha=0.02$')
plt.ylim(0, 1.0)
plt.xlabel('times attended')
plt.ylabel('weight')
plt.axvline(x=100, color='k', linestyle='--', alpha=0.2)

plt.plot(w_ball, label='$w_{ball}$')
plt.plot(w_flower, label='$w_{flower}$')
# plt.plot(w_baby, label='$w_{baby}$')
plt.legend()
```
