```python
from state import State

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
app_state = State()

w_ball = []
w_flower = []
w_baby = []
ALPHA = 0.010

for i in range(300):
    context = app_state.context
    toys = context.toys
    ball = next((x for x in toys if x.name == ''), None)
    w_ball.append(ball.weight)

    flower = next((x for x in toys if x.name == ''), None)
    w_flower.append(flower.weight)

    baby = next((x for x in toys if x.name == ''), None)
    w_baby.append(baby.weight)

    if i < 100:
        context.on_visual_attention_3('', alpha=ALPHA)
    elif i < 200:
        context.on_visual_attention_3('', alpha=ALPHA)
    elif i < 300:
        context.on_visual_attention_3('', alpha=ALPHA)

plt.figure(figsize=(5,3), dpi=150, facecolor='white')
plt.title(f'$\\alpha={ALPHA}$')
plt.ylim(-0.05, 1.05)
plt.xlabel('times attended')
plt.ylabel('weight')
plt.axvline(x=100, color='k', linestyle='--', alpha=0.4)
plt.axvline(x=200, color='k', linestyle='--', alpha=0.4)

plt.plot(w_ball, label='$w_{ball}$')
plt.plot(w_flower, label='$w_{flower}$')
plt.plot(w_baby, label='$w_{baby}$', linestyle='dotted')
plt.legend()
```
