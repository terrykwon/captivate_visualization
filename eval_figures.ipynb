```python
import pandas as pd 
import numpy as np 

from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Helvetica']

import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Total number of responses in 10 minutes

def plot_response_counts():
    counts_app = np.array([143, 99, 47])
    counts_cards = np.array([121, 111, 51])

    mean_app = np.mean(counts_app)
    mean_cards = np.mean(counts_cards)
    std_app = np.std(counts_app)
    std_cards = np.std(counts_cards)

    labels = ['Control', 'Captivate!']
    x_pos = np.arange(len(labels))
    means = [mean_cards, mean_app]
    stds = [std_cards, std_app]

    fig, ax = plt.subplots(figsize=(4,3), dpi=150)
    bars = ax.bar(x_pos, means, yerr=stds, align='center', alpha=0.5, ecolor='black', capsize=10)
    bars[0].set(color='gray', hatch='/')
    bars[1].set(color='blue', hatch='.')
    ax.set_ylabel('Number of responsive utterances')
    ax.set_xticks(x_pos)
    ax.set_xticklabels(labels)
    ax.yaxis.grid(True)

    # Save the figure and show
    plt.tight_layout()
    plt.savefig('outputs/results_response_counts.png')
    plt.show()

plot_response_counts()
 
```
